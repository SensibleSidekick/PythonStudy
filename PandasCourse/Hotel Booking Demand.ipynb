```python
#Dealing with Categorical (Non number) features.
#creating a new dataframe with jsut the categorical features from our original dataframe
cat_cols = [col for col in data_df.columns if data_df[col].dtype == "str"]
cat_cols
```

```python
#country and hotel features
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#Label encode country
cat_df["country"] = le.fit_transform(cat_df["country"])

#Label encode hotel
cat_df["hotel"] = le.fit_transform(cat_df["hotel"])
```

```python
#one-hot encoding 
cat_df = pd.get_dummies(data = cat_df, columns = ["meal", "market_segment", "distribution_channel", "reserved_room_type", "assigned_room_type", "customer_type", "deposit_type", "reservation_status"])
cat_df.drop("res_weekday", inplace=True, axis=1)
```

```python
#Numerical dataframe
num_df = data_df.drop(columns= cat_cols)
```

```python
num_df
```

```python
final_df = pd.concat([num_df, cat_df], axis=1)
```

```python
final_df
```

```python
#Correlation analysis
#create a correlation matrix
corr_matrix = final_df.corr().abs()
corr_matrix
```

```python
#setting the threshold
threshold = 0.85
```

```python
import numpy as np
#This is taking the matrix and removing the diagonal cells and the triangle of cells below that since the diagonals
#are always going to be one since they are comparing to themselves, and the cells below that are a mirror of the
#cells above it. This sets these cells to NaN.
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
#ensure only 1 of highly correlated features get dropped. 
to_drop = []
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i,j]) >= threshold:
            colname = corr_matrix.columns[i]
            to_drop.append(colname)
```

```python
to_drop
```

```python
#Drop those columns
final_df.drop(columns=to_drop, inplace=True)
```

```python
#should we drop agent feature?
final_df.agent.isnull().sum()
```

```python
#ideally look up more info about these agents. In this case null is "not applicable". Keep column.
#imputing with 0

final_df["agent"] = final_df["agent"].fillna(0)
```
